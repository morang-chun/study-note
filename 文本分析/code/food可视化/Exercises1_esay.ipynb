# 1、绘制柱状图

## 题目要求：请根据给出的FoodsData.csv文件，将其中的名称作为横轴，钙的含量作为纵轴，绘制出柱状图，并正常显示中文刻度

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus']=False

%matplotlib inline
```

```python
data = pd.read_csv("./FoodsData(1).csv")
```

```python
data.head()
```

```python
# 将其中的名称作为横轴，钙的含量作为纵轴，绘制出柱状图，并正常显示中文刻度
plt.figure(figsize=(30,10),dpi=80)
y = data["钙"]
x = data["名称"]
plt.bar(x,y,width=0.4,color="pink")
plt.xlabel("食物名称",fontsize=20)
plt.ylabel("钙的含量",fontsize=20)
plt.title("食物中钙的含量柱状图",fontsize=30)
plt.show()
```

# 4、绘制饼图

## 题目要求：根据给出的 FoodsData.csv数据，绘制出饼图，要求体现出大分类列每个分类占比总分类的比例是多少

```python
data.info()
```

```python
# plt.pie(x, labels, autopct, shadow,startangle)
# 根据给出的 FoodsData.csv数据，绘制出饼图，要求体现出大分类列每个分类占比总分类的比例是多少
x = data["大分类"].value_counts()
```

```python
plt.figure(figsize=(15,9))
labels = ["水果 / 蔬菜 / 豆类 / 坚果","焙烤食品","乳与乳制品","肉及肉制品","粮食和粮食制品","水产及其制品","冷冻饮品","可可制品 / 巧克力 / 糖果","饮料类","甜味料，包括蜂蜜","酒类"]
plt.pie(x,labels=labels,autopct='%.f%%',startangle=0)


plt.title("大分类列每个分类占比总分类的比例",fontsize=20)
plt.show()
```
