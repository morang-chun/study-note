# 1、绘制柱状图

```python
from matplotlib import pyplot as plt
import pandas as pd

# 获取数据
df = pd.read_csv('FoodsData.csv')
df.head()
# 提取绘图需要的数据
x = df['名称']
y = df['钙']


# # 以下另种配置中文字体的方式根据自己电脑系统进行匹配
# MAC端设置中文字体
plt.rcParams['font.family'] = 'Arial Unicode MS'

# windows端配置中文字体
# plt.rcdefaults()
# plt.rcParams['font.sans-serif'] = ['SimHei'] 


plt.figure(figsize=(20,8),dpi=80)
figure = plt.bar(x, y, )
for i in figure:
    height = i.get_height()
    x = i.get_x()
    width = i.get_width()
    plt.text(x+width/2, height+0.1, height, ha='center')

plt.show()
```

# 2、绘制直方图

```python
from matplotlib import pyplot as plt
import random

list1 = [random.randint(117,427) for i in range(30)]

# 设置组距
distance = 10
# 计算组数
group_num = int((max(list1) - min(list1)) / distance)
# 绘制直方图
plt.hist(list1, bins=group_num, color='yellow', edgecolor='black')
# 展示图形
plt.show()
```

# 3、绘制散点图

```python
from matplotlib import pyplot as plt
import pandas as pd

df = pd.read_csv('FoodsData.csv')
x = df['卡路里']
y = df['脂肪']

plt.figure(figsize=(20,8),dpi=80)
plt.scatter(x, y, s=50, c='red', marker='1',alpha=0.7, linewidths=2)
plt.show()
```

# 4、绘制饼图

```python
from matplotlib import pyplot as plt
import pandas as pd


# # 以下另种配置中文字体的方式根据自己电脑系统进行匹配
# MAC端设置中文字体
plt.rcParams['font.family'] = 'Arial Unicode MS'

# windows端配置中文字体
# plt.rcdefaults()
# plt.rcParams['font.sans-serif'] = ['SimHei'] 

df = pd.read_csv('FoodsData.csv')
# 分组
groups = df.groupby(by='大分类').size()
# 绘制饼图
plt.pie(groups,labels=groups.index, autopct='%.1f%%',startangle=90)
# 图形展示
plt.show()
```
