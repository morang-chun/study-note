### 1.导入pandas库和matplotlib库下的pyplot方法

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### 2.通过plt.rcParams['font.sans-serif'] = ['SimHei']方法设置字体

```python
plt.rcParams['font.sans-serif']=['SimHei']
```

### 3.创建出如下所示双折线图
#### ——要求：
#### 1.双折线要在图中展示
#### 2.在图中展示图例
#### 3.设置x轴为“第一季度”、“第二季度”、“第三季度”和“第四季度”
#### 4.设置xlabel为“季度值（三个月一季度）”
#### 5.设置ylabel为'收入(万元)'
#### 6.设置title为“季度收入/投入折线图”

```python
data1 = pd.DataFrame({
    "收入":[400,590,500,450],
    "投入":[200,470,250,300]
}, index=['第一季度','第二季度','第三季度','第四季度'])
```

```python
data1.head()
```

```python

plt.figure(figsize = (10,8),dpi = 80)
data1.plot()
plt.xlabel("季度值（三个月一个季）")
plt.ylabel("收入/投入（万元）")
plt.title("季度收入/投入折线图",size=20)
plt.show()
```
