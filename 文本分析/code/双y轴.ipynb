        董事长         80        35000
        刘总           90        22000
        马副总         95        18000
        张总监         83        20000
        董事长儿子     97        30000
        技术总监       98        25000

 #### 2 请根据以上数据,绘制出一张表述员工职位与当前月薪关系的柱状图,要求横轴为员工职位,纵轴为当前月薪  
 
 #### 3 请根据以上数据,绘制出一张表述员工职位分别与学习成绩和当前月薪的双y轴图形,图形有折线图与柱状图组成,要求展示出完整的图例信

## 首先定义员工数据信息

```python
names = ['董事长','刘总','马副总','张总监','董事长儿子','技术总监']
scores = [80,90,95,83,70,91]
salary = [35000,22000,18000,20000,30000,25000]
```

## 导入相关包

```python

import matplotlib.pyplot as plt
```

## 根据姓名与以前的学习成绩绘制折线图

```python

plt.rcdefaults()

plt.rcParams['font.sans-serif'] = ['SimHei'] 
# 指定默认字体,固定用法,也可导入本地字体库的方式来设置字体 windows系统用此句

#plt.rcParams['font.sans-serif'] = ['Arial Unicode MS'] 
#mac系统用此句

plt.figure(figsize=(7,4),dpi=80)
plt.plot(names,scores,label='学习成绩',color='green')
plt.legend()
plt.xlabel('姓名')
plt.ylabel('学习成绩')
plt.grid()
plt.show()
```

## 根据员工姓名与员工当前薪资绘制柱状图

```python
plt.figure(figsize=(7,4),dpi=80)
plt.bar(names,salary,label='当前薪资',color='red',width=0.5,alpha=0.3)
plt.legend()
plt.xlabel('姓名')
plt.ylabel('当前薪资')
plt.grid()
plt.show()
```

## 根据姓名与学习成绩以及当前薪资绘制双y轴图形,以姓名作为共享轴,本题重点在于设置不重叠的图例,使用了跟小课不一样的方法(get_legend_handles_labels方法)进行绘制,该方法可以直接返回当前图形的图形对象以及图例对象,非常容易的就能得到图形图例对象,也作为扩展共大家学习

```python
#设置画布大小
fig = plt.figure(figsize=(8,4),dpi=80)
#添加坐标系对象
ax = fig.add_subplot(111)
#绘制折线图,并设置折线相应的状态
ax.plot(names,scores,label='学习成绩',marker='o',markeredgecolor = 'blue')
#在ax坐标系上制作双y轴对象ax1坐标系
ax1 = ax.twinx()
#通过ax1坐标系绘制柱状图
ax1.bar(names,salary,label='当前薪资',width=0.35,color='red',alpha=0.3)
#get_legend_handles_labels方法可以获取当前坐标系相应图形的图形对象以及该图形对象的实例,如下获取ax坐标系上的折线对象以及它的图例对象,
# 其中hander_1代表折线图形对象,label_1代表该折线图形对象的图例
handler_1,label_1 = ax.get_legend_handles_labels()
#如下获取ax1坐标系的图例对象: 柱状图对象(handler_2),以及柱状图的图例对象(label_2)
handler_2,label_2 = ax1.get_legend_handles_labels()
# 合并图例,并使用legend方法显示图例
plt.legend(handler_1+handler_2,label_1+label_2)
#设置ax以及ax1两者共享坐标轴(横轴)的名称为:  姓名
ax.set_xlabel('姓名')
#设置ax坐标系y轴数据的名字
ax.set_ylabel('学习成绩(折线图)')
#设置ax1坐标系y轴数据的名字
ax1.set_ylabel('当前薪资(柱状图)')
plt.grid()
plt.show()
```
