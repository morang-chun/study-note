   ### &emsp;&emsp;&emsp;员工姓名     &emsp;学习成绩    &emsp; 当前月薪
        董事长         80        35000
        刘总           90        22000

        张总监         83        20000
        董事长儿子     97        22000

 #### 1 请根据以上数据,绘制出一张表述员工职位与学习成绩关系的折线图,要求横轴为员工职位,纵轴为学习成绩  
 
 #### 2 请根据以上数据,绘制出一张表述员工职位与当前月薪关系的柱状图,要求横轴为员工职位,纵轴为当前月薪  
 
 #### 3 请根据以上数据,绘制出一张表述员工职位分别与学习成绩和当前月薪的双y轴图形,图形有折线图与柱状图组成,要求展示出完整的图例信

## 首先定义员工数据信息

```python
names = ['董事长','刘总','马副总','张总监','董事长儿子','技术总监']
scores = [80,90,95,83,70,91]
salary = [35000,22000,18000,20000,30000,25000]
```

## 导入相关包

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

## 设置中文字体,并根据姓名与以前的学习成绩绘制折线图

```python
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus']=False
```

```python
plt.figure(figsize=(10,6),dpi=80)
plt.plot(names,scores,label="score",color="red")
plt.legend()
plt.xlabel("姓名")
plt.ylabel("学习成绩")
plt.gca()
plt.show()
```

## 根据员工姓名与员工当前薪资绘制柱状图

```python
# plt.rcdefaults()
plt.figure(figsize=(10,9),dpi=80)
plt.bar(names,salary,label="当前薪资",width=0.4,color=["pink"])#,"black","blue","red","green","brown"]
plt.legend()
plt.xlabel("姓名",fontsize=20)
plt.ylabel("薪资",fontsize=20)
plt.title("员工姓名与员工当前薪资柱状图",fontsize=20)
plt.gca()

for x,y in enumerate(salary):
#     print(x,y)
    plt.text(x,y+300,y,ha='center',fontsize=18)
    
plt.grid()
plt.show()
```

## 根据姓名与学习成绩以及当前薪资绘制双y轴图形,以姓名作为共享轴,本题重点在于设置不重叠的图例,可以查询get_legend_handles_labels方法的用法,该方法可以直接返回当前图形的图形对象以及图例对象,非常容易的就能得到图形图例对象(答案中有详细的注释),也作为扩展共大家学习

```python
plt.rcdefaults()
plt.rcParams['font.sans-serif'] = 'SimHei'
fig = plt.figure(figsize=(8,6),dpi=80)
dy = fig.add_subplot(111)
dy.plot(names,scores,label="学习成绩",marker='o',color="blue")

dy1 = dy.twinx()
dy1.bar(names,salary,label="当前薪资",width=0.35,color='brown',alpha=0.5)
for x,y in enumerate(salary):
#     print(x,y)
    plt.text(x,y+300,y,ha='center',fontsize=10)

handler_1,label_1 = dy.get_legend_handles_labels()
handler_2,label_2 = dy1.get_legend_handles_labels()
plt.legend(handler_1+handler_2,label_1+label_2)
plt.xlabel("姓名")
dy.set_ylabel("学习成绩",fontsize=16)
dy1.set_ylabel("当前薪资",fontsize=16)
plt.grid()
plt.show()
```
